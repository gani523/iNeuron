```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap
from sklearn import decomposition
from sklearn import datasets
```

```python
#loading the dataset
iris=datasets.load_iris()
print("Independent variables sample data:\n",iris.data[:5,:])
print("Feature names:\n",iris.feature_names)
print("Dependent variable:\n",iris.target_names)
```

```python
X = iris["data"][:, (1,2, 3)] 
y = iris["target"] 

#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)

#Reducing the dimensions for 3D
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
X = pca.fit_transform(X)

fig = plt.figure(1, figsize=(5, 5))
#plt.clf() means clear current figure
ax = Axes3D(fig)
#plt.cla() means clear current axis
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=ListedColormap(['Green','Blue','Red']), edgecolor='k')
plt.show()
```
